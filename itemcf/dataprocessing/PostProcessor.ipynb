```python
popularity_file = open('foundation_popularity.txt', 'w')
similarity_file = open('foundation_similarity.txt', 'w')
file = open('coOccurrenceMatrix.txt', 'r')
lines = file.readlines()
for line in lines:
    lst = line.split('\t')
    foundationA = lst[0].split(':')[0]
    foundationB = lst[0].split(':')[1]
    if foundationA == foundationB:
        print (foundationA.strip() + '\t' + lst[1])
        popularity_file.write(foundationA.strip() + '\t' + lst[1])
    else:
        similarity_file.write(foundationA.strip() + '\t' + foundationB.strip() + '\t' + lst[1])
file.close()
popularity_file.close()
similarity_file.close()
```
